```python
import gradio as gr
from transformers import pipeline

# Load summarization model
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Load question-answering model
question_answerer = pipeline("question-answering", model="deepset/roberta-base-squad2")

def chatbot_interface(content, choice, query=None):
    if choice == "Summarize":
        # Summarize content
        summary = summarizer(content, max_length=150, min_length=30, do_sample=False)
        return summary[0]['summary_text']
    elif choice == "Answer Question":
        # Answer question based on content
        if query:
            result = question_answerer(question=query, context=content)
            return result['answer']
        else:
            return "Please provide a question."
    else:
        return "Invalid choice. Please select 'Summarize' or 'Answer Question'."

# Define the Gradio interface
def chatbot(content, choice, query=""):
    return chatbot_interface(content, choice, query)

# Create Gradio inputs and outputs
content_input = gr.Textbox(label="Enter Content", lines=5, placeholder="Paste or type your content here...")
choice_input = gr.Radio(["Summarize", "Answer Question"], label="Choose Action")
query_input = gr.Textbox(label="Enter Question (if 'Answer Question' is selected)", placeholder="Type your question here...")
output_text = gr.Textbox(label="Output", placeholder="The result will be displayed here.")

# Build the interface
interface = gr.Interface(
    fn=chatbot,
    inputs=[content_input, choice_input, query_input],
    outputs=output_text,
    title="AI Chatbot",
    description="Provide content and choose an action: summarize the content or answer questions based on it."
)

# Launch the Gradio app
interface.launch()
```
